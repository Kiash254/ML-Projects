## Data Collection and Preprocessing

```python
#Load the credit card datasets and check the first five rows of the datasets
credit=pd.read_csv('/content/creditcard.csv')
#Check the first five rows
credit.head()
```

* this dataset is highly imbalanced where i will need to do some data augmentation

0-Normal Transactions
1-fraudelent transactions

```python
#separate the data for analysis
legit=credit[credit.Class==0]
fraud=credit[credit.Class==1]
```

* i will use undersample  to make sure the data is equal distributed where one class of legit transactiona and fraud


```python
legit_sample=legit.sample(n=492)
```

```python
#Concatonate two DataFrames
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

## Separate the data into Features and Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into trainginf and testing data for model training

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

# Model training and Evaluation

```python
model=LogisticRegression()
```

```python
#Fit the model
model.fit(X_train,Y_train)
```

## Model Evaluation of the model

```python
train_test=model.predict(X_train)
acc=metrics.accuracy_score(train_test,Y_train)
print("The Accuracy of the model is:",acc)
```

```python
test_data=model.predict(X_test)
acc=metrics.accuracy_score(test_data,Y_test)
print("The Accuracy of the test data of the model is:",acc)
```
