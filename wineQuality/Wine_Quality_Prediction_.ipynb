## Future selection and enginneering

```python
  X=wine.drop('quality',axis=1)
```

```python
#Binary of the data
Y=wine['quality'].apply(lambda y_value:1 if y_value >=7 else 0)
```

```python
print(Y)
```

```python
#split the data to test and train
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

## Random forest

image.png

```python
model=RandomForestClassifier()
```

```python
model.fit(X_train,Y_train)
```

## Model Evaluation

```python
train_test=model.predict(X_train)
acc=accuracy_score(train_test,Y_train)
print("The Accuracy of the model is:",acc)
```

```python
train_test=model.predict(X_test)
acc=accuracy_score(train_test,Y_test)
print("The Accuracy of the model is:",acc)
```

## Building a  predictive system where we will make a system a chemical parameters values and check the Quality of the wine

```python
input_data=(7.9,0.6,0.06,1.6,0.069,15.0,59.0,0.9964,3.3,0.46,9.4)

#Change the input data in to numpy array
data_tup=np.asarray(input_data)

#Reshape the data into
data_res=data_tup.reshape(1,-1)
predict=model.predict(data_res)

print(predict)

if (predict[0] == 1):
  print("Good Quality Wine")

else:
  print("Hello Please it a bad Quality one")
```
